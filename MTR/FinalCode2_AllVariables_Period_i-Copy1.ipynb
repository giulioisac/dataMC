```python
qr_mc1.correctAllY(X, ylist, quantiles,EBEE="EB", relativePath= "weights_Period_"+period) 
qr_mc2.correctAllY(X, ylist, quantiles,EBEE="EE", relativePath= "weights_Period_"+period) 
dfmcEB= qr_mc1.getDF() 
dfmcEE= qr_mc2.getDF() 
```
