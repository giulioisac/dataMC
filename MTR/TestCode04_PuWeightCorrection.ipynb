```python
#histogramming wrt rho
rv  = np.arange(0.0,40.0,0.4)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel("rho")
    
#compute histograms:
mc_bins, _edges = np.histogram(qr_mc_ConWeights.df.loc[:,"rho"], rv)
edges = _edges[:len(_edges)-1]
data_bins, _edges = np.histogram(dfdata["rho"], rv)
edges = _edges[:len(_edges)-1]
#calculate error (no need to normalize in height since same size):
bincenters = 0.5*(_edges[1:]+_edges[:-1])
width      = 0
menStd     = np.sqrt(mc_bins)
menStddata     = np.sqrt(data_bins)
    
#plot
plt.errorbar(bincenters, data_bins, yerr=menStddata, color='r', capsize=0)
plt.errorbar(bincenters, mc_bins,yerr=menStd,  color='g',capsize=0)
    
ax.set_title('Rho histogram')
```

```python
#take the ratio
#do the ratio
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel("rho")

h1=np.true_divide(mc_bins,data_bins)

#propagate relative 
h1er= menStd/data_bins

#set cutoff
cutoff=5
for i in range(len(h1)):
    if (h1[i]>cutoff): 
        h1[i]=cutoff
        h1er[i]=0.1
        
#print h1.size, h1er    
    #plot
plt.errorbar(edges,h1, yerr=h1er,capsize=0,color="g")# uncorrected
ax.set_ylim([0, 10])
    
ax.set_title('ratio with cutoff')
```

```python
#find the correction for each weight:
import pandas as pd
#find corrected value of correction:
qr_mc_ConWeights.df['bin'] = np.digitize(qr_mc_ConWeights.df['rho'],edges)

#compute correction vector--> why the minus 1?
qr_mc_ConWeights.df["corr"]= h1[qr_mc_ConWeights.df['bin'].values-1]

#correct the weights:
qr_mc_ConWeights.df["corrweight"]=qr_mc_ConWeights.df["weight"].values/qr_mc_ConWeights.df["puweight"].values*qr_mc_ConWeights.df["corr"].values
```
